```python
import os
from py2neo import Graph,Node, Relationship
from dotenv import load_dotenv
load_dotenv()
# API 相关 (优先从 .env 读取)
KEY = os.getenv("DATABASE_KEY")
```

```python
try:
    graph = Graph("bolt://localhost:7687", auth=("neo4j", KEY))
    # 运行一个简单的查询测试
    result = graph.run("RETURN '连接成功！' as message").data()
    print(result[0]['message'])
except Exception as e:
    print(f"连接失败: {e}")
```

```python
# 假设你之前提取到的三元组数据如下：
triplets = [
    {"head": "GraphRAG", "relation": "使用", "tail": "Neo4j"},
    {"head": "Neo4j", "relation": "存储在", "tail": "D盘"},
    {"head": "Python", "relation": "连接", "tail": "Neo4j"}
]
```

```python
item = triplets[0]
head_node = Node("Concept", name=item["head"])
tail_node = Node("Concept", name=item["tail"])
rel = Relationship(head_node, item["relation"], tail_node) # 只是将两个节点连接在了一起，并没有存入数据库中（还在内存中）
```

```python
graph.merge(head_node, "Concept", "name")
graph.merge(tail_node, "Concept", "name")
graph.merge(rel)
```
